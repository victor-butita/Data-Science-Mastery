```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
os.chdir(r'C:\Users\Pc\Desktop\Data Science Notes\IRIS')
```

```python
cars_data = pd.read_csv('Toyota.csv',index_col=0,na_values=['??','????'])
```

```python
cars_data.dropna(axis = 0, inplace = True)
```

## Scatter Plot

```python
sns.set(style="darkgrid")
```

```python
sns.regplot(x=cars_data['Age'],y=cars_data['Price'])
```

### fit_reg=True by default, fit_reg shows the regression line

```python
sns.regplot(x=cars_data['Age'],y=cars_data['Price'],fit_reg=False) 
```

### marker = * , used to change the marker

```python
sns.regplot(x=cars_data['Age'],y=cars_data['Price'],fit_reg=False, marker='*') 
```

## lmplot
###### *legend is by default True, you can change it to False...*

```python
sns.lmplot(x='Age', y='Price', data=cars_data, fit_reg=False, hue='FuelType', palette='Set1', legend=True)
#legend is by default True
```

### Histogram

```python
sns.distplot(cars_data['Age'], kde=False) #kde=True by default
```

```python
sns.distplot(cars_data['Age'], kde=False, bins=5) #limit the number of Bins
```

### Bar plot

```python
%%time
sns.countplot(x='FuelType', data=cars_data)
```

### Grouped bar plot

```python
sns.countplot(x='FuelType', hue='Automatic', data=cars_data)
```

### Box and whiskers plot

```python
sns.boxplot(x=cars_data['FuelType'],y=cars_data['Price'])
```

### Grouped box and whiskers plot

```python
sns.boxplot(x='FuelType',y='Price',hue='Automatic',data=cars_data)
```

### Box-whiskers plot and Histogram

```python
f,(ax_box,ax_hist) = plt.subplots(2,gridspec_kw={"height_ratios":(.20,.80)})
sns.boxplot(cars_data['Price'], ax=ax_box)
sns.distplot(cars_data['Price'], ax=ax_hist, kde=False)
```

### Pairwise Plot

```python
sns.pairplot(cars_data,kind='scatter', hue='FuelType')
plt.show()
```
